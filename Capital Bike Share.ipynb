```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
day = pd.read_csv('day.csv')
day['dteday'] = day['dteday'].astype('datetime64')
day['dteday'] = day['dteday'].apply(lambda x: x.year)
month = {1: 'January', 2: 'February', 3:'March', 4:'April', 5: 'May', 6:'June', 7:'July',8:'August',9:'September',10:'October',11:'November', 12:'December'}
day['mnth'] = day['mnth'].apply(lambda x: month[x])
season = {1: 'Spring', 2:'Summer', 3:'Fall', 4:'Winter'}
day['season'] = day['season'].apply(lambda x: season[x])
weekday = {0:'Sunday', 1:'Monday', 2:'Tuesday', 3:'Wednesday', 4:'Thursday', 5:'Friday', 6:'Saturday'}
day['weekday'] = day['weekday'].apply(lambda x: weekday[x])
day
```

```python
#1b
month_casual = pd.DataFrame(zip(day['mnth'],day['casual']), columns=['Month', 'Casual Bikers'])
month_casual = month_casual.groupby('Month').sum().sort_values(by=['Casual Bikers'], ascending=False).head(3)
ax = sns.barplot(month_casual.index, month_casual['Casual Bikers'],ci=None)
ax.set_title('Top Three Month Totals For Casual Bikers')
month_casual
```

```python
#1a
month_registered = pd.DataFrame(zip(day['mnth'],day['registered']), columns=['Month', 'Registered Bikers'])
month_registered = month_registered.groupby('Month').sum().sort_values(by=['Registered Bikers'], ascending=False).head(3)
ax = sns.barplot(month_registered.index, month_registered['Registered Bikers'],ci=None)
ax.set_title('Top Three Month Totals For Registered Bikers')
month_registered
```

```python
#1c
seasonal_registered = pd.DataFrame(zip(day['season'], day['registered']))
seasonal_registered.columns = ['Season','Registered Bikers']
seasonal_registered = seasonal_registered.groupby("Season").sum()
seasonal_registered
ax = sns.barplot(seasonal_registered.index, seasonal_registered['Registered Bikers'],ci=None)
ax.set_title('Total Number of Registered Bikers')
seasonal_registered
```

```python
#1d
seasonal_casual = pd.DataFrame(zip(day['season'], day['casual']))
seasonal_casual.columns = ['Season','Casual Bikers']
seasonal_casual = seasonal_casual.groupby("Season").sum()
ax = sns.barplot(seasonal_casual.index, seasonal_casual['Casual Bikers'],ci=None)
ax.set_title('Total Number of Casual Bikers By Season')
seasonal_casual
```

```python
year_2011 = pd.DataFrame(zip(day[(day.dteday == 2011)]['mnth'], day[(day.dteday == 2011)]['cnt']))
year_2011.columns = ['Month','Total']
year_2012 = pd.DataFrame(zip(day[(day.dteday == 2012)]['mnth'], day[(day.dteday == 2012)]['cnt']))
year_2012.columns = ['Month','Total']
```

```python
#1e
year_2011_avg = year_2011.groupby('Month').mean()
year_2011_avg = year_2011_avg['Total'].astype('int64')
year_2011_avg = pd.DataFrame(year_2011_avg).rename(columns = {'Total':'Average Count'})
year_2011_avg
ax = sns.barplot(year_2011_avg.index, year_2011_avg['Average Count'],ci=None)
ax.set_title('Average Count of Bicycles By Month for 2011')
ax.set_xticklabels(ax.get_xticklabels(),rotation=60)
year_2011_avg
```

```python
year_2012_avg = year_2012.groupby('Month').mean()
year_2012_avg = year_2012_avg['Total'].astype('int64')
year_2012_avg = pd.DataFrame(year_2012_avg).rename(columns = {'Total':'Average Count'})
year_2012_avg
ax = sns.barplot(year_2012_avg.index, year_2012_avg['Average Count'],ci=None)
ax.set_title('Average Count of Bicycles By Month for 2012')
ax.set_xticklabels(ax.get_xticklabels(),rotation=60)
year_2012_avg
```

```python
hour = pd.read_csv('hour.csv')
hour['dteday'] = hour['dteday'].astype('datetime64')
hour['dteday'] = hour['dteday'].apply(lambda x: x.year)
month = {1: 'January', 2: 'February', 3:'March', 4:'April', 5: 'May', 6:'June', 7:'July',8:'August',9:'September',10:'October',11:'November', 12:'December'}
hour['mnth'] = hour['mnth'].apply(lambda x: month[x])
season = {1: 'Spring', 2:'Summer', 3:'Fall', 4:'Winter'}
hour['season'] = hour['season'].apply(lambda x: season[x])
hour['weekday'] = hour['weekday'].apply(lambda x: weekday[x])
hour
```

```python
year_cnt = pd.DataFrame(zip(hour['dteday'], hour['cnt']))
year_cnt.columns = ['Year', 'Count']
year_cnt = year_cnt.groupby('Year').sum()
ax = sns.barplot(year_cnt.index, year_cnt['Count'],ci=None)
ax.set_title('Number of Bike Rentals For Each Year')
year_cnt
```

```python
year2012_temp = pd.DataFrame(zip(hour[(hour.dteday == 2012)]['mnth'], hour[(hour.dteday == 2012)]['temp'] ))
year2012_temp.columns = ['Month', 'Temperature']
year2012_temp['Temperature'] = year2012_temp['Temperature'] * 47 -8
year2012_temp_avg = year2012_temp.groupby('Month').mean()
ax = sns.barplot(year2012_temp_avg.index, year2012_temp_avg['Temperature'],ci=None)
ax.set_title('Average Temperature For Each Month of 2012')
ax.set_xticklabels(ax.get_xticklabels(),rotation=60)
year2012_temp_avg
```

```python
year2012_hum = pd.DataFrame(zip(hour[(hour.dteday == 2012)]['mnth'], hour[(hour.dteday == 2012)]['hum']*100))
year2012_hum.columns = ['Month','Humidity']
year2012_hum_avg = year2012_hum.groupby('Month').mean()
year2012_hum_avg.index[0]
months = month.values()
hums = []
for i in range(len(months)):
    for j in range(len(year2012_hum_avg.index)):
        if year2012_hum_avg.index[j] == months[i]:
            hums.append(year2012_hum_avg['Humidity'][j])
        #print(hums)
year2012_hum_avg = pd.DataFrame(zip(months, hums))

# year2012_hum_avg.index = month.values()
year2012_hum_avg.columns = ['Month','Humidity']
ax = sns.barplot(year2012_hum_avg['Month'], year2012_hum_avg['Humidity'],ci=None)
ax.set_title('Average Humidity For Each Month of 2012')
ax.set_xticklabels(ax.get_xticklabels(),rotation=60)
year2012_hum_avg
```

```python
work_reg = pd.DataFrame(zip(hour[(hour.workingday == 1)]['weekday'], hour[(hour.workingday == 1)]['registered']))
work_reg.columns = ['Weekday','Registered Bikes']
work_reg_total = work_reg.groupby('Weekday').sum()
work_reg_total = work_reg_total.sort_values(by='Registered Bikes', ascending=False)
ax = sns.barplot(work_reg_total.index, work_reg_total['Registered Bikes'],ci=None)
ax.set_title('Total Number of Registered Bikes By Weekday')
ax.set_xticklabels(ax.get_xticklabels(),rotation=60)
work_reg_total
```

```python
hour_reg = pd.DataFrame(zip(hour['hr'],hour['registered']))
hour_reg.columns = ['Hour','Registered Bikes']
hour_reg_total = hour_reg.groupby('Hour').sum()
hour_reg_total = hour_reg_total.sort_values(by='Registered Bikes', ascending=False)
ax = sns.barplot(hour_reg_total.index, hour_reg_total['Registered Bikes'],ci=None)
ax.set_title('Total Number of Registered Bikes By Hour')
ax.set_xticklabels(ax.get_xticklabels(),rotation=60)
hour_reg_total
```

```python
work_casual = pd.DataFrame(zip(hour[(hour.workingday == 1)]['weekday'], hour[(hour.workingday == 1)]['casual']))
work_casual.columns = ['Weekday','Casual Bikes']
work_casual_total = work_casual.groupby('Weekday').sum()
work_casual_total = work_casual_total.sort_values(by='Casual Bikes', ascending=False)
ax = sns.barplot(work_casual_total.index, work_casual_total['Casual Bikes'],ci=None)
ax.set_title('Total Number of Casual Bikes By Weekday')
ax.set_xticklabels(ax.get_xticklabels(),rotation=60)
work_casual_total
```
